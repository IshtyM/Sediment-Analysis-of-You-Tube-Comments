```python
#to analize the polarity of all comments

polarity=[]
for i in GB_Comments['comment_text']:
    Pole=TextBlob(i).sentiment.polarity
    polarity.append(Pole)
GB_Comments['Polarity']=polarity
```

```python
GB_Comments.head(20)
```

```python
GB_Comments_Positive_Polarity=GB_Comments[GB_Comments ['Polarity'] >0 ]
print("New data contains only positive polarity having head ")
GB_Comments_Positive_Polarity
```

```python
print("The new data contains", GB_Comments_Positive_Polarity.shape , "rows and columns respectively")
```

```python
!pip install wordcloud
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
print("The stopwords are used by library as ")
stopwords
```

```python
Total_Comments = (' '.join(GB_Comments_Positive_Polarity["comment_text"])).lower()
```

```python
wordcloud = WordCloud(width=1000, height=500, stopwords=stopwords).generate(Total_Comments)
```

```python
print("The plot of wordcloud is ")
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis("off")
```

```python
Total=(' '.join(GB_Comments_Positive_Polarity["comment_text"])).split()
a=len(Total)
print("The total number of words in positive polarity comments are ")
```

```python
import collections
Common_words = [word for word in Total_Comments.split() if word in stopwords]
Count_Common_Words = collections.Counter(Common_words)

Common = {}

for letter, count in Count_Common_Words.most_common(100):
    Common[letter] = count
    
print("The count of top 100 letters common from stopwords are ")
    
for i,j in Common.items():
        print('Word: {0}, count: {1}'.format(i,j))
```

```python
Common_words = [word for word in Total_Comments.split() if word in stopwords]
Count_Common_Words = collections.Counter(Common_words)

Common = {}

for letter, count in Count_Common_Words.most_common(100):
    Common[letter] = (count/a)*100
    
print("The percentage of 100 letters coomon from stopwords are ")
    
for i,j in Common.items():
        print('Word: {0}, count: {1}'.format(i,j))
```

```python
Common_words = [word for word in Total_Comments.split() if word not in stopwords]
Count_Common_Words = collections.Counter(Common_words)

Common = {}

for letter, count in Count_Common_Words.most_common(100):
    Common[letter] = count
    
print("The count of top 100 letters which are not in stopwords are ")
    
for i,j in Common.items():
        print('Word: {0}, count: {1}'.format(i,j))
```
